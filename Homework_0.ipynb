```python
import cv2

image = cv2.imread('Data/Apollo_11_launch.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
import matplotlib.pyplot as plt

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
cropped_image = image[100:600, 300:900]

plt.imshow(cropped_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
dark_image = cv2.subtract(cropped_image, 100)

plt.imshow(dark_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
rotated_image = cv2.rotate(dark_image, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(rotated_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
cv2.imwrite('Data/Processed_Apollo_11_launch.jpg', rotated_image)
```
